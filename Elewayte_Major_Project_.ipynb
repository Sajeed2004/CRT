```python
import pandas as pd

url = "https://raw.githubusercontent.com/arora123/Data/master/emp-data.csv"
df = pd.read_csv(url)

display(df.head())
```

```python
gender_counts = df['Gender'].value_counts()
print("Total number of Males and Females:")
display(gender_counts)
```

```python
print("\nGender distribution by Department:")
display(df.groupby('Department')['Gender'].value_counts())

print("\nGender distribution by Location:")
display(df.groupby('Loc')['Gender'].value_counts())
```

```python
df['Salary'] = df['Salary'].replace({'\$': '', ',': ''}, regex=True).astype(float)

average_salary_by_department = df.groupby('Department')['Salary'].mean().sort_values(ascending=False)

print("\nAverage Salary by Department:")
display(average_salary_by_department)

highest_avg_salary_department = average_salary_by_department.index[0]
print(f"\nDepartment with the highest average salary: {highest_avg_salary_department}")
```

```python
average_salary_by_location = df.groupby('Loc')['Salary'].mean().sort_values(ascending=False)

print("\nAverage Salary by Location:")
display(average_salary_by_location)

highest_avg_salary_location = average_salary_by_location.index[0]
print(f"\nLocation with the highest average salary: {highest_avg_salary_location}")
```

```python
rating_counts = df['Rating'].value_counts()
total_employees = len(df)

rating_percentage = (rating_counts / total_employees) * 100

print("\nPercentage of employees by Rating:")
display(rating_percentage)
```

```python
gender_salary_by_department = df.groupby(['Department', 'Gender'])['Salary'].mean().unstack()
gender_salary_by_department['Pay Gap'] = gender_salary_by_department['Male'] - gender_salary_by_department['Female']

print("\nGender Pay Gap by Department:")
display(gender_salary_by_department)
```

```python
gender_salary_by_location = df.groupby(['Loc', 'Gender'])['Salary'].mean().unstack()
gender_salary_by_location['Pay Gap'] = gender_salary_by_location['Male'] - gender_salary_by_location['Female']

print("\nGender Pay Gap by Location:")
display(gender_salary_by_location)
```
